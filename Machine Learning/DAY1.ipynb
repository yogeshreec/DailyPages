```python
h=img.shape[0]
w=img.shape[1]
r=np.min((h,w))//2
print(r)
cx=w//2
cy=h//2
print(f"Radius={r},Centerx={cx},Centery={cy}")
```

```python
spotimg=img.copy()
for x in range(w):
    for y in range(h):
        d= np.sqrt((x-cx)**2+(y-cy)**2)  #d-> distance
        if d>r:spotimg[y,x]=255  #
plt.imshow(spotimg);
```
